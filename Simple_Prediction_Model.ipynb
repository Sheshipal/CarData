```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/output_numerical.csv", index_col=0)
```

```python
data.head()
```

```python
data.info
```

```python
Y = data["price in rupee"]
X = data[['owner', 'driven km', 'transmission', 'fuel', 'body type', 'engine cc',
       'max power bhp', 'wheel size inches', 'no of seats', 'mileage kmpl', 'purchase year', 'company']]

print(X.shape)
print(Y.shape)
```

```python
x_train,x_test, y_train,y_test = train_test_split(X, Y, test_size = 0.30, shuffle=True)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
linear_regression = LinearRegression()
linear_regression.fit(x_train, y_train)
```

```python
y_pred = linear_regression.predict(x_test)
```

```python
print(y_test[0:5])
print(y_pred[0:5])
```

```python
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```

```python
import keras
import tensorflow as tf
```

```python
model = keras.Sequential([
    keras.layers.Dense(32, activation=tf.nn.relu, input_shape=[1,12]),
    keras.layers.Dense(32, activation=tf.nn.relu),
    keras.layers.Dense(16, activation=tf.nn.relu),
    keras.layers.Dense(8, activation=tf.nn.relu),
    keras.layers.Dense(1)
])
```

```python
optimizer = tf.keras.optimizers.RMSprop(0.0099)
model.compile(loss='mean_squared_error',optimizer=optimizer)
model.fit(x_train,y_train,epochs=500)
```

```python
y_pred = model.predict(x_test)
```

```python
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
```
